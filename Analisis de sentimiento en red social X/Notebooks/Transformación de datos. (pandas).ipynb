```python
df = pd.read_json('tweets_crudos_laroja.json', lines=True) # importamos los datos crudos
print(df.columns)
```

```python
pd.set_option('display.max_columns', None) # desplegamos todas las columnas
pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', 0) 
pd.set_option('display.max_rows', None) # desplegamos todas las filas

print(df.head())
```

```python
# 2Ô∏è‚É£ Seleccionar columnas √∫tiles
df = df[['created_at', 'text', 'author_id']]
df.head(70)
```

```python
import re

# 3Ô∏è‚É£. limpiamos las filas de la columna texto, para as√≠ obtener una columna mas legible

def limpiar_texto(text):
    text = re.sub(r"http\S+", "", text)           # elimina URLs
    text = re.sub(r"@\w+", "", text)              # elimina menciones
    text = re.sub(r"#\w+", "", text)              # elimina hashtags
    text = re.sub(r"\n", " ", text)               # reemplaza saltos de l√≠nea por espacio
    text = re.sub(r"[^A-Za-z√Å√â√ç√ì√ö√ë√°√©√≠√≥√∫√±0-9\s]", "", text)  # elimina s√≠mbolos raros, deja letras y n√∫meros
    return text.strip()

df['tweet_limpio'] = df['text'].apply(limpiar_texto)
df.head(10)
```

```python
# 4Ô∏è‚É£ Formatear fecha y eliminar duplicados / vac√≠os

df['created_at'] = pd.to_datetime(df['created_at'])
df = df.drop_duplicates(subset='tweet_limpio')
df = df.dropna(subset=['tweet_limpio'])
```

```python
df['created_at'].dt.date.value_counts().sort_index()
```

```python
# 5Ô∏è‚É£ Reordenar columnas
df = df[['created_at', 'author_id', 'tweet_limpio']]
df.head()
```

```python
# 6Ô∏è‚É£ Guardar limpio y ordenado
df.to_csv('tweets_limpios_ordenados.csv', index=False, encoding='utf-8-sig')
df.to_parquet('tweets_limpios_ordenados.parquet', index=False)

print("‚úÖ Archivos guardados:")
print(" - tweets_limpios_ordenados.csv (para Excel)")
print(" - tweets_limpios_ordenados.parquet (formato eficiente para an√°lisis)")
```

```python
# 7Ô∏è‚É£ Analisis de sentimiento con HuggingFace
from transformers import pipeline
clasificador = pipeline("sentiment-analysis", model= "pysentimiento/robertuito-sentiment-analysis")

# Creamos 3 columnas con el resultado de cada texto del twwet, el sentimiento y la confianza

df['resultado'] = df['tweet_limpio'].apply(lambda x: clasificador(x[:512])[0])
df['sentimiento'] = df['resultado'].apply(lambda x: x['label'])
df['confianza'] = df['resultado'].apply(lambda x: x['score'])
df.head(50)
```

```python
# Agregamos polaridad a la confianza
def convertir_polaridad(fila):
    if fila['sentimiento'] == 'POS':
        return fila['confianza']
    elif fila['sentimiento'] == 'NEG':
        return -fila['confianza']
    else:
        return 0

df['confianza_polarizada'] = df.apply(convertir_polaridad, axis=1)
# Ordenamos las columnas
df = df[['created_at', 'author_id', 'tweet_limpio', 'sentimiento', 'confianza_polarizada']]
df.head()
```

```python
# Eliminamos time zone de la columna created_at 
df['created_at'] = df['created_at'].dt.tz_localize(None)
df.head()
```

```python
# Exportamos a un archivo excel los datos limpios para reporte

df.to_excel('tweets_limpios2.xlsx', index=False)
```
